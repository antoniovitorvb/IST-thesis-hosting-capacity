```python
import requests
import pandas as pd
```

```python
data_json = requests.get('https://e-redes.opendatasoft.com/api/explore/v2.1/catalog/datasets/3-consumos-faturados-por-municipio-ultimos-10-anos/records?limit=-1').json()
# data_json

consumo_pt = pd.DataFrame(data_json['results'])
consumo_pt
```
